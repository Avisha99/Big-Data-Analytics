## Add new columns with labels "mean" and "date" to the DataFrame df. 

## The two columns are as follows: 

- ### Column "mean": mean of the 12 windspeeds recorded at 12 locations each day

- ### Column "date": date in the format 19YY-MM-DD. this column is derived from the first three columns of the dataframe df. you will need the Pandas function to_datetime() 

```python
df['mean']=df[['Loc1','Loc2','Loc3','Loc4','Loc5','Loc6','Loc7','Loc8','Loc9','Loc10','Loc11','Loc12']].mean(axis=1)

df.head()
```

```python
df["Year"]=df["Year"]+1900
df['date']=pd.to_datetime(df[['Year','Month','Day']])
df.head()
```

## Generate two subplots. 

- ### Subplot 1 contains mean weekly windspeed at Loc 2 smoothened using rolling average window of 3 weeks.

- ### Subplot 2 contains median monthly windspeed at Loc 2 smoothened using rolling average window of 4 months.

```python
plt.figure(figsize=(35,10))
#Axis 1
axOne=plt.subplot(1,2,1)
plt.title("Mean weekly Windspeed at Loc2", fontsize=30)
df['Loc2'].rolling(21).mean().plot(fontsize=20,c="blue")

#Axis 2
axTwo=plt.subplot(1,2,2)
plt.title("Median weekly Windspeed at Loc2", fontsize=30)
df['Loc2'].rolling(120).median().plot(fontsize=20,c="red")

plt.show()
```
