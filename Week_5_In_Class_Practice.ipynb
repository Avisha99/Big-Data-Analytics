```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy
import statistics as stats
from scipy.stats import norm
```

```python
df=pd.read_csv("NonlinearPor_Perm_data.csv")
df.head()
```

```python
PorosityA=df['PorosityA'].values
PermeabilityA = df['LogPermA'].values
PorosityB = df['PorosityB'].values
PermeabilityB = df['LogPermB'].values
```

```python
plt.figure(figsize = (20,10))

# creating the first axis 
axOne = plt.subplot(1,2,1)
plt.scatter(PorosityA,PermeabilityA,c = 'blue',alpha=0.2, s = 20,edgecolors = 'black')
plt.xlabel('Porosity (fraction)'); plt.ylabel('LogPermeability (mD)')
plt.title("Permeability vs. Porosity for formation A", fontsize = 30);plt.grid()

# creating the second axis
axTwo = plt.subplot(1,2,2)
plt.scatter(PorosityB,PermeabilityB,c = 'blue',alpha=0.2, s = 20,edgecolors = 'black')
plt.xlabel('Porosity (fraction)'); plt.ylabel('LogPermeability (mD)')
plt.title("Permeability vs. Porosity for formation B", fontsize = 30);plt.grid()
plt.show()
```

```python
 # Pearson product moment correlation
    
from scipy.stats import pearsonr         

corr, corr_p_value = pearsonr(PorosityA,PermeabilityA)

print('Sample Pearson product moment correlation coefficient of permeability and porosity in  formation A is ' + str(round(corr,3)))
print('Sample Pearson product moment correlation coefficient of permeability and porosity p-value in  formation A is ' + str(round(corr_p_value,20)))
```

```python
corr_p_value
```

```python
 # Pearson product moment correlation
    
from scipy.stats import pearsonr         

corr, corr_p_value = pearsonr(PorosityB,PermeabilityB)

print('Sample Pearson product moment correlation coefficient of permeability and porosity in  formation B is ' + str(round(corr,3)))
print('Sample Pearson product moment correlation coefficient of permeability and porosity p-value in  formation B is ' + str(round(corr_p_value,20)))
```

```python
corr_p_value
```

```python
# On comparison of pearson correlation coefficient of formation A and formation B, which is 0.791 and 0.982 respectively.
# on the scatter plot, most of the data points are in the same line which means a linear relationship. 
# Therefore, porosity and permeability are strongly correlated in both the formations whereas in the case of formation B,the value of correlation coefficient is much more greater which is a bit unrealistic.
```

```python
import seaborn as sns

plt.figure(figsize = (30,20))

# creating the first axis 
axOne = plt.subplot(1,2,1)
sns.regplot(PorosityA, PermeabilityA, ci=None)
plt.xlabel('Porosity (fraction)'); plt.ylabel('LogPermeability (mD)')
plt.title("Permeability vs. Porosity for formation A", fontsize = 35);plt.grid()

# creating the second axis
axTwo = plt.subplot(1,2,2)
sns.regplot(PorosityB, PermeabilityB, ci=None)
plt.xlabel('Porosity (fraction)'); plt.ylabel('LogPermeability (mD)')
plt.title("Permeability vs. Porosity for formation B", fontsize = 35);plt.grid()
plt.show()
```

```python
# As seen from the graphs above, a linear correlation between permeability and porosity can be determined for both the formations.
# However, the relation for formation A is not as much reliable as it is for formation B because for one value of porosity, there are a range of values of permeability. 
# In the case of formation B, the correlation is reliable.
```

```python
# spearman rank correlation

from scipy.stats import spearmanr         

rank_corr = spearmanr(PorosityA,PermeabilityA)[0]
rank_corr_p_value = spearmanr(PorosityA,PermeabilityA)[1]

print('Sample Spearman rank correlation coefficient of permeability and porosity is in formation A ' + str(round(rank_corr,2)))
print('Sample Spearman rank correlation coefficient of permeability and porosity p-value is in formation A is ' + str(round(rank_corr_p_value,4)))
```

```python
corr_p_value
```

```python
 # spearman rank correlation
    
from scipy.stats import spearmanr        

rank_corr = spearmanr(PorosityB,PermeabilityB)[0]
rank_corr_p_value = spearmanr(PorosityB,PermeabilityB)[1]

print('Sample Spearman rank correlation coefficient of permeability and porosity is in formation B ' + str(round(rank_corr,2)))
print('Sample Spearman rank correlation coefficient of permeability and porosity p-value is in formation B is ' + str(round(rank_corr_p_value,20)))
```
