<span id = "cm"><h3>Creating Correlation Matrix</h3></span>

<p> Pandas has a corr method which accepts the argument methods which defines the correlation coefficents returned. e.g. Pearson, Spearman</p>

<p>Correlation coefficients will only be returned for numeric data</p>

```python
wellA1.head()
```

```python
wellA1.set_index('DATEPRD',inplace=True) # make date the index
wellA1.head()
```

```python
# Get correlation matrix for Well 15/9 F-1 C

# Dropping date as they would not provide relevant information in relation to correlation

corr_matrix = wellA1.corr(method = "spearman")

# index is not included in the correlation computation
```

```python
corr_matrix
```

```python
# Removing the repeated section of the correlation matrix
mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

mask
```

```python
# Create Axis
fig, ax = plt.subplots(figsize=(12, 12))

# Using Seaborn HeatMap
sns.heatmap(corr_matrix, mask = mask, square = True, linewidths = .8, cbar_kws = {'shrink': .4,  'ticks' : [-1, -.5, 0, 0.5, 1]},
            cmap = 'coolwarm', vmin = -1,  vmax = 1, annot = True, annot_kws = {'size': 14})

# Setting Axis Properties
ax.set_yticklabels(corr_matrix.columns, rotation = 0, size = 15)
ax.set_xticklabels(corr_matrix.columns, rotation = 90, size =15)
ax.set_title("Correlation Matrix", size = 25)

sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```
