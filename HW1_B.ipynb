## What is the Cumulative Production Per Year ? plot it.

```python
well1["year"]= well1.DATEPRD.dt.year
well1["month"]=well1.DATEPRD.dt.month
```

```python
fig, ax= plt.subplots(figsize=(20,10))
#making plot using pandas
well1[["BORE_OIL_VOL","year"]].groupby(["year"]).sum().plot(ax=ax,kind="bar", fontsize=20)

#setting plot properties
ax.set_ylabel("Total Oil Volume", fontsize=20)
ax.set_xlabel("Year", fontsize=20)
ax.set_title("Total Production Per Year From Well 15/9-F-11", fontsize=20)
plt.show()
```

## Draw a line Plot of Oil and Production History on same axes

```python
fig, ax=plt.subplots(figsize=(30,20))
#creating twin axis from existing axis
ax1=ax.twinx()
#plotting oil Volume on axis 1
ax.plot(well1["DATEPRD"],well1["BORE_OIL_VOL"], c="blue", label="Oil", linewidth=3)

#plotting gas volume on axis 2
ax1.plot(well1["DATEPRD"],well1["BORE_GAS_VOL"], c="red", label="GAS", linewidth=3)

#Setting properties for axis 1
ax.legend(fontsize=30, loc=2)
ax.tick_params(axis='both', which='major', labelsize=30)
ax.set_ylabel("Oil Production", fontsize= 40)

#setting properties for axis 2
ax1.legend(fontsize=30,loc=1)
ax1.tick_params(axis='both', which='major', labelsize=30)
ax1.set_ylabel("Gas Production (m3)", fontsize= 40)

#setting title plot
plt.title('Production History', fontsize=35)
plt.show()
```

## Using box, plots what is the distribution of Pressure/Temperature by year ?

```python
well1['press/temp']=well1['AVG_DOWNHOLE_PRESSURE']/well1['AVG_DOWNHOLE_TEMPERATURE']
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="year", y="press/temp", data=well1)
plt.title("AVERAGE DOWNHOLE PRESSURE/TEMPERATURE BY YEAR using Seaborn", fontsize=35)
```
